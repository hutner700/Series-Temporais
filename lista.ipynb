```python
exp = pd.read_excel("Exportações.xls",0)
exp.columns = ["Data", "Valor"]
exp['Data'] = exp['Data'].astype(str)
exp.set_index("Data",inplace=True)
```

```python
inf = pd.read_excel("Inflação.xls",0)
inf.columns = ["Data", "Valor"]
inf['Data'] = inf['Data'].astype(str)
inf.set_index("Data",inplace=True)
```

```python
taxaJ = pd.read_excel("Taxa de Juros.xls",0)
taxaJ.columns = ["Data", "Valor"]
taxaJ['Data'] = taxaJ['Data'].astype(str)
taxaJ.set_index("Data",inplace=True)
```

```python
def plotar(dataframe, nome_y):
    plt.plot(dataframe['Valor'])
    plt.title(nome_y)
    plt.xlabel('Ano')
    plt.xticks(dataframe.index[::120], dataframe.index[::120])
    plt.show()
```

```python
plotar(exp, "Exportações")
```

```python
plotar(inf, "Inflação")
```

```python
plotar(taxaJ, "Taxa de Juros")
```

```python
# Para a série exp
plot_acf(exp['Valor'], lags=36, title="ACF exportação")
plot_pacf(exp['Valor'], lags=36, title="PACF exportação")
```
